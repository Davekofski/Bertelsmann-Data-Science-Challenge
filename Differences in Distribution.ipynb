```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from scipy.stats import norm
import seaborn as sns
import random
```

```python
#specify the parameters of the experiment
sample_size=1000
sample_space=[1,2,3,4]
sample=[]
```

```python
#now draw the random numbers and add them to the sample list

for i in range(1,sample_size):
    sample.append(random.choice(sample_space))
```

```python
#plot the distribution of the samples

sns.countplot(sample)
plt.ylabel("observed draws")
plt.title("1000 random draws from [1,2,3,4]")
plt.xlabel("number observed from draw")
plt.show()
```

```python
#draw the distribution of samples from normal distribution with same parameters

sns.distplot(np.random.normal(loc=np.mean(sample_space), scale=np.std(sample_space), size=sample_size),kde=False)
plt.ylabel("observed draws")
plt.title("1000 random draws from normal with mean 2.5 and std.dev. 1.25")
plt.xlabel("number observed from draw")
plt.show()
```
