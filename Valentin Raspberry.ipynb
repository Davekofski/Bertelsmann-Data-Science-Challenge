```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
#airqual=pd.read_table("bme360-AQ.txt",sep=":")
airqual=pd.read_csv("airqual.csv",sep=";")
```

```python
def edit(X):
    return X.split()[0]
```

```python
airqual["Gas"]=airqual["Gas"].apply(edit)
airqual["Humidity"]=airqual["Humidity"].apply(edit)
airqual["Humidity"]=airqual["Humidity"].apply(float)
airqual["Gas"]=airqual["Gas"].apply(float)
```
